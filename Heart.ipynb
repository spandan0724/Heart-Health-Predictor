Next specific code from scikit-learn is imported:

-'accuracy score': A function for evaluating classification accuracy.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

The read_csv function from pandas to read the 'heart.csv' file into a DataFrame named heart

```python
heart = pd.read_csv('heart.csv')
heart
```

1 = Defective Heart
2 = Healthy Heart

```python
x = heart.drop(columns = 'target', axis=1)
y = heart['target']
```

```python
x
```

```python
y
```

Splitting Into Training Data and Testing Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, stratify=y, random_state=2)
```

```python
print(x.shape, xtrain.shape, xtest.shape )
```

Model Training using Logistic Regression

```python
Model = LogisticRegression()
```

```python
Model.fit(xtrain, ytrain)
```

Model Evaluation

```python
#accuracy score
x_train_prediction = Model.predict(xtrain)
training_data_accuracy = accuracy_score(x_train_prediction, ytrain)
```

```python
training_data_accuracy
```

```python
#accuracy score
x_test_prediction = Model.predict(xtest)
testing_data_accuracy = accuracy_score(x_test_prediction, ytest)
```

```python
testing_data_accuracy
```

Building A Predictive System

```python
input_data = (60,  1,	0,	125,	258,	0,	0,	141,	1,	2.8,	1,	1,	3)
inputdata_toarray = np.asarray(input_data)
inputdata_reshaped = inputdata_toarray.reshape(1,-1)
prediction = Model.predict(inputdata_reshaped)
print(prediction)
if prediction == '0':
    print("The person has heart diease")
else:
    print("The person doesnot have any heart diease")    
```
